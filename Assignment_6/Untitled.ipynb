```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
M = float(input("Enter a value for M: "))
```

```python
x = np.linspace(-10, 10, 200)
```

```python
y1 = M * x**2
y2 = M * np.sin(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='y = M·x²', color='blue', linestyle='--')
plt.plot(x, y2, label='y = M·sin(x)', color='green', linestyle='-')
plt.title('Mathematical Functions: y = M·x² and y = M·sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

**Q2**

```python
subjects = ['Math', 'Science', 'English', 'History', 'Art']
scores = [85, 90, 78, 88, 92]
data = pd.DataFrame({'Subject': subjects, 'Score': scores})
print(data)
```

```python
plt.figure(figsize=(8, 5))
barplot = sns.barplot(x='Subject', y='Score', data=data, palette='muted')
```

```python
for i, score in enumerate(scores):
    barplot.text(i, score + 1, str(score), ha='center', va='bottom', fontweight='bold')

plt.title('Scores by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y')
```

**Q4**

```python
data = pd.read_csv('company_sales_data.csv')
```

```python
print(data.head())
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=data, marker='o', color='blue')

plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(data['month_number'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

for product in product_columns:
    sns.lineplot(x='month_number', y=product, data=data, label=product, marker='o')

plt.title('Sales of All Products Over Months')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.xticks(data['month_number'])
plt.legend(title='Product')
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(16, 12))
axes = axes.flatten()

# Skip the first column (month_number) and loop through the rest
for i, column in enumerate(data.columns[1:]):  
    sns.barplot(x='month_number', y=column, data=data, ax=axes[i], palette='Set2')
    axes[i].set_title(f'{column.capitalize()} Over Months')
    axes[i].set_xlabel('Month')
    axes[i].set_ylabel(column.capitalize())
    axes[i].set_xticks(data['month_number'])

plt.tight_layout()
plt.show()
```
